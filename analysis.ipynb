```python
import pandas as pd
```

```python
df=pd.read_csv("data/Amazon_Sale_Report.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Unnamed: 22', 'fulfilled-by'])
```

```python
df['currency']=df['currency'].fillna('INR')
```

```python
df['promotion-ids']=df['promotion-ids'].fillna('No promotion')
```

```python
multiple = ['ship-city', 'ship-state', 'ship-postal-code', 'ship-country']
df[multiple] = df[multiple].fillna('Unknown')
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.head()
```

```python
df.columns
```

```python
df.to_csv("data/Amazon_Sale_Report1.csv", index=False)
```

```python
df.loc[df['status'] == 'Cancelled', 'courier_status'] = 'Cancelled'
df.loc[df['status'] == 'Shipped - Delivered to Buyer', 'courier_status'] = 'Shipped'
df.loc[df['status'] == 'Shipped - Returned to Seller', 'courier_status'] = 'Shipped'
```

```python
df.amount=df.amount.fillna("unknown")
```
